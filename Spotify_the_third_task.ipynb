```python
Unnamed: 0
رقم تسلسلي أو مؤشر للسطر، وليس له معنى تحليلي مباشر.
Track.Name
اسم الأغنية.
Artist.Name
اسم الفنان أو الفرقة الموسيقية التي غنت الأغنية.
Genre
نوع الأغنية الموسيقي (مثال: pop، reggaeton، rap).
Beats.Per.Minute
عدد الإيقاعات في الدقيقة (BPM)، وهو مؤشر على سرعة الأغنية.
Energy
مستوى الطاقة في الأغنية (من 0 إلى 100) يعني أغنية أكثر حيوية ونشاطًا.
Danceability
مدى توافق الأغنية مع الرقص (من 0 إلى 100) تعني إيقاعها مناسب للرقص.
Loudness..dB..
مستوى الصوت (الشدة) بالأ ديسيبل (dB)، كلما كان الرقم أقرب إلى الصفر كانت الأغنية أعلى صوتًا.
Liveness
احتمالية أن الأغنية تم تسجيلها مباشرة (live) يعني أن لها طابع عروض حية.
Valence.
مشاعر السعادة أو الإيجابية في  = أغاني أكثر فرحًا وإيجابية.
Length.
طول الأغنية بالثواني.
Acousticness..
مدى اعتماد الأغنية على الآلات الصوتية بدون  = أكثر صوتاً طبيعياً.
Speechiness.
مدى وجود كلمات منطوقة في الأغنية (مثل الراب) = فيها أكثر من حديث أو راب.
Popularity
مؤشر على شهرة الأغنية (من 0 إلى 100) = أكثر شيوعًا أو استماعًا.
```
